```python
# 1. Rozbalení tvých dat z cloudu (vykřičník na začátku znamená, že to je příkaz pro linuxový terminál Colabu)
!unzip -q dataset_cleaned.zip -d /content/

# 2. Import knihoven pro strojové učení
import tensorflow as tf
from tensorflow.keras import layers, models

# 3. Načtení dat a jejich rozdělení
# Neuronová síť se musí na něčem učit (80 % dat) a na zbytku se testuje (20 % dat), zda se to naučila dobře
dataset_cesta = '/content/dataset_cleaned'

train_data = tf.keras.preprocessing.image_dataset_from_directory(
    dataset_cesta,
    validation_split=0.2,
    subset="training",
    seed=123,
    image_size=(128, 128), # Naše data už takto mají 128x128, ale Keras to pro jistotu ověří
    batch_size=32 # Model se bude učit po dávkách 32 fotek
)

val_data = tf.keras.preprocessing.image_dataset_from_directory(
    dataset_cesta,
    validation_split=0.2,
    subset="validation",
    seed=123,
    image_size=(128, 128),
    batch_size=32
)

# 4. Architektura Konvoluční neuronové sítě (CNN)
model = models.Sequential([
    # Vstupní vrstva + Normalizace (převod barev pixelů z 0-255 na 0-1)
    layers.InputLayer(shape=(128, 128, 3)),
    layers.Rescaling(1./255),

    # První konvoluce (AI zde hledá základní hrany, stíny, obrysy)
    layers.Conv2D(32, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)), # Zmenšení obrázku pro rychlejší výpočet

    # Druhá konvoluce (AI skládá obrysy do tvarů - oči, uši, nos)
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),

    # Třetí konvoluce (AI už hledá specifické rysy Ronalda)
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),

    # Převedení 2D obrázku do 1D řady čísel (zploštění)
    layers.Flatten(),
    layers.Dense(64, activation='relu'),

    # Výstupní vrstva s 1 neuronem (funkce Sigmoid vrací číslo od 0.0 do 1.0)
    # 0 = Ostatní, 1 = Ronaldo
    layers.Dense(1, activation='sigmoid')
])

# 5. Kompilace modelu (nastavení toho, jak se má model "trestat" za chyby)
model.compile(optimizer='adam',
              loss='binary_crossentropy', # Používáme, protože máme jen 2 třídy (Ronaldo vs Není Ronaldo)
              metrics=['accuracy'])

# 6. Spuštění tréninku na 10 "epoch" (10x projde celý dataset)
print("\nZačínám trénovat model...")
history = model.fit(train_data, validation_data=val_data, epochs=10)

# 7. Uložení natrénovaného "mozku" do souboru
model.save('ronaldo_model.h5')
print("\nHotovo! Model je úspěšně uložen jako ronaldo_model.h5")
```
